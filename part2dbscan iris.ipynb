```python
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
x=iris.data
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=4)
dbscan.fit(x)
labels=dbscan.labels_
```

```python
np.histogram(labels)
```

```python
plt.hist(labels)
```
