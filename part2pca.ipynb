```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=plt.figure()
ax=Axes3D(fig)
ax=fig.add_subplot(111,projection="3d")
#versicolor=iris.data[50:100]
#versicolor=versicolor[:,[0,1,3]]
ax.scatter(iris.data[:,0],iris.data[:,1],iris.data[:,2],c='red')
ax.set_title('iris')
ax.set_xlabel('sepal length')
ax.set_ylabel('sepal width')
ax.set_zlabel('petal width')
plt.show()
```

```python
vc_pca=PCA()
vc_pca.fit(iris.data)
vc_pca_transformed=vc_pca.transform(iris.data)
```

```python
nfeaturs=range(vc_pca.n_components_)
var_featcurs=vc_pca.explained_variance_
```

```python
plt.bar(nfeaturs,var_featcurs)
plt.xlabel('pca featcure')
plt.ylabel('variance')
plt.xticks(nfeaturs)
plt.show()
```

```python
dim_r=PCA(n_components=2)
dim_r.fit(iris.data)
dim_r_transformed=dim_r.transform(iris.data)
dim_r_transformed
```

```python
plt.scatter(dim_r_transformed[:,0],dim_r_transformed[:,1],c=iris.target)
plt.show()
```
