```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
seed=pd.read_csv('seeds-width-vs-length.csv')
seed
```

```python
seed=seed.to_numpy()
correlaton,_=pearsonr(seed[:,0],seed[:,1])
correlaton
```

```python
plt.plot(seed[:,0],seed[:,1],'o')
plt.xlabel('width')
plt.ylabel('lenght')
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
pca.fit(seed)
transformed=pca.transform(seed)
```

```python
mean=pca.mean_
mean
```

```python
#factor proportionof corrected variance
fpc=pca.components_[0]
fpc
```

```python
plt.scatter(seed[:,0],seed[:,1])
plt.arrow(mean[0],mean[1],fpc[0],fpc[1],color='magenta',width=0.02)
plt.show()
```

```python
plt.scatter(transformed[:,0],transformed[:,1])
plt.show()
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris_model=PCA()
iris_model.fit(iris.data[:,[0,2]])
iris_transformed=iris_model.transform(iris.data[:,[0,2]])
mean=iris_model.mean_
fpc=iris_model.components_[0]
```

```python
plt.scatter(iris.data[:,0],iris.data[:,2])
plt.arrow(mean[0],mean[1],fpc[0],fpc[1],color='red',width=0.1)
plt.show()
```

```python
plt.scatter(iris_transformed[:,0],iris_transformed[:,1])
plt.show()
```
