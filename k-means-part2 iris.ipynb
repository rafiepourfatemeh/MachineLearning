```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
x=iris.data
x
```

```python
kmn=KMeans(n_clusters=3)
kmn.fit(x)
labels=kmn.predict(x)
labels
```

```python
center=kmn.cluster_centers_
center
```

```python
plt.scatter(x[:,0],x[:,2],c=labels)
plt.scatter(center[:,0],center[:,2],marker='x',s=150)
plt.show()
```

```python
kmn.inertia_
```

```python
y=[]
for k in np.arange(1,6):
    kmn=KMeans(n_clusters=k)
    kmn.fit(x)
    y.append(kmn.inertia_)
y
```

```python
plt.plot(np.arange(1,6),y,'o-')
plt.xlabel('number of clusters')
plt.ylabel('inertia')
plt.show()
```

```python
import pandas as pd
```

```python
kmn=KMeans(n_clusters=3)
kmn.fit(x)
labels=kmn.predict(x)
labels
```

```python
clusters =pd.DataFrame(x.copy())
clusters['Cluster_Prediction'] = kmn.fit_predict(x)
clusters.sample(50)
```

```python
fig, ax = plt.subplots(figsize=(15,7)) 
plt.scatter(x=clusters[clusters['Cluster_Prediction'] == 0][0],
            y=clusters[clusters['Cluster_Prediction'] == 0][1],
            s=70,edgecolor='teal', linewidth=0.3, c='teal', label='Iris-versicolor')


plt.scatter(x=clusters[clusters['Cluster_Prediction'] == 1][0],
            y=clusters[clusters['Cluster_Prediction'] == 1][1],
            s=70,edgecolor='lime', linewidth=0.3, c='lime', label='Iris-setosa')


plt.scatter(x=clusters[clusters['Cluster_Prediction'] == 2][0],
            y=clusters[clusters['Cluster_Prediction'] == 2][1],
            s=70,edgecolor='magenta', linewidth=0.3, c='magenta', label='Iris-virginica')

plt.scatter(x=center[:, 0], y=center[:, 1], s = 170, c = 'yellow', label = 'Centroids',
                                   edgecolor='black', linewidth=0.3)
plt.legend(loc='upper right')
plt.xlim(4,8)
plt.ylim(1.8,4.5)
ax.set_ylabel('Sepal Width (in cm)')
ax.set_xlabel('Sepal Length (in cm)')
plt.title('Clusters', fontsize = 20)
plt.show()
```
