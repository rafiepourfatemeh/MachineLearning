```python
import pandas as pd
import numpy as np
mydata=pd.read_csv('customer.csv')
mydata
```

```python
mydata.dtypes
```

```python
gender=set(mydata.Gender)#change type gender
Gender={}
for i ,j in enumerate(gender,start=1):
    x={j:i}
    Gender.update(x)
mydata=mydata.applymap(lambda s:Gender.get(s) if s in Gender else s)
mydata
```

```python
mydata=mydata.set_index('CustomerID')
mydata
```

```python
mydata.isnull().sum()
```

```python
mydata.shape
```

```python
mydata.drop_duplicates(inplace=True)
mydata.shape
```

```python
x=np.asarray(mydata)
x
```

```python
from sklearn.preprocessing import scale,normalize,minmax_scale
#x=minmax_scale(x,feature_range=(-1,1))
#x=normalize(x,norm='l2',axis=0)
x=scale(x)
x
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x[:, 1], x[:, 3], marker='.')
```

```python
from sklearn.cluster import KMeans
```

```python
kmn=KMeans(n_clusters=2)
kmn.fit(x)
labels=kmn.predict(x)
labels
```

```python
center=kmn.cluster_centers_
```

```python

plt.scatter(x[:,1],x[:,3],c=labels)
plt.scatter(center[:,1],center[:,3],marker='x',s=150)
plt.legend()
#for i, label in enumerate(labels):
    #plt.text(x[:,1][i],x[:,3][i],label)
plt.show()
```

```python
kmn.inertia_
```

```python
y=[]
for k in np.arange(1,10):
    kmn=KMeans(n_clusters=k)
    kmn.fit(x)
    y.append(kmn.inertia_)
y
```

```python
plt.plot(np.arange(1,10),y,'o-')
plt.xlabel('number of clusters')
plt.ylabel('inertia')
plt.show()
```

```python
from sklearn.cluster import MeanShift
```

```python
ms=MeanShift(bandwidth=2)
ms.fit(x)
labels=ms.labels_
center=ms.cluster_centers_
print(labels)
```

```python
plt.scatter(x[:,1],x[:,3],c=labels)
plt.scatter(center[:,1],center[:,3],marker='x',s=300,linewidths=5)
for i, label in enumerate(labels):
    plt.text(x[:,1][i],x[:,3][i],label)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=7)
dbscan.fit(x)
labels=dbscan.labels_
labels
```

```python
plt.scatter(x[:,1],x[:,2],c=labels)
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage,dendrogram,fcluster
```

```python
hirarachical=linkage(x,method='complete')
dendrogram(hirarachical)
plt.show()
```

```python
labels=fcluster(hirarachical,4,criterion='distance')
labels
```

```python
plt.scatter(x[:,1],x[:,2],c=labels)
plt.show()
```
